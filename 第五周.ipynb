```python
# [Week 4] Transformer 時間序列預測 - 聯發科 (2454.TW)

# [1] 環境設置
!pip install yfinance

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import tensorflow as tf
import os
import random
import warnings
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras import layers, models, Input

warnings.filterwarnings('ignore')

# 設定隨機種子
RANDOM_SEED = 42
def set_random_seeds(seed=RANDOM_SEED):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    os.environ['TF_DETERMINISTIC_OPS'] = '1'
    try:
        tf.config.experimental.enable_op_determinism()
    except AttributeError:
        pass

set_random_seeds(RANDOM_SEED)

# 字體設定
!wget -O TaipeiSansTCBeta-Regular.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download
import matplotlib.font_manager as fm
fm.fontManager.addfont('TaipeiSansTCBeta-Regular.ttf')
plt.rc('font', family='Taipei Sans TC Beta')

# [2] 資料獲取
def get_stock_data(stock_id='2454.TW', start_date='2020-01-01', end_date='2024-10-10'):
    print(f" 下載 {stock_id} 資料...")
    start = dt.datetime.strptime(start_date, '%Y-%m-%d')
    end = dt.datetime.strptime(end_date, '%Y-%m-%d')
    df = yf.download(stock_id, start=start, end=end)
    df = df.reset_index()
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)

    # 清理 MultiIndex
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.droplevel('Ticker')

    return df

# 獲取聯發科資料
stock_data = get_stock_data('2454.TW')

# 繪製原始股價
plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, stock_data['Close'], label='Close Price', color='blue')
plt.title('聯發科 (2454) 股價走勢')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# [3] 特徵工程 (OHLCV)
def prepare_features(data):
    # 選取 Open, High, Low, Close, Volume
    feature_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
    df = data[feature_cols].copy()
    df = df.dropna()
    return df, feature_cols

df_features, feature_names = prepare_features(stock_data)

# [4] 資料前處理 (正規化 & 視窗切割)
WINDOW_SIZE = 20
target_col_name = 'Close'

# 分割訓練/測試集
train_size = int(len(df_features) * 0.8)
train_data = df_features[:train_size]
test_data = df_features[train_size:]

# 正規化
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train_data)
test_scaled = scaler.transform(test_data)

# 建立滑動視窗
def create_dataset(dataset, window_size, target_col_idx):
    X, y = [], []
    for i in range(len(dataset) - window_size):
        X.append(dataset[i:i+window_size, :])
        y.append(dataset[i+window_size, target_col_idx])
    return np.array(X), np.array(y)

# 取得目標欄位索引
target_idx = df_features.columns.get_loc(target_col_name)

X_train, y_train = create_dataset(train_scaled, WINDOW_SIZE, target_idx)
X_test, y_test = create_dataset(test_scaled, WINDOW_SIZE, target_idx)

print(f" 訓練集形狀: {X_train.shape}")
print(f" 測試集形狀: {X_test.shape}")

# [5] Transformer 模型建構
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Normalization and Attention
    x = layers.LayerNormalization(epsilon=1e-6)(inputs)
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(x, x)
    x = layers.Dropout(dropout)(x)
    res = x + inputs

    # Feed Forward Part
    x = layers.LayerNormalization(epsilon=1e-6)(res)
    x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    return x + res

def build_transformer_model(input_shape, head_size=256, num_heads=4, ff_dim=4, num_transformer_blocks=4, mlp_units=[128], dropout=0.4, mlp_dropout=0.4):
    inputs = Input(shape=input_shape)
    x = inputs

    # 堆疊 Transformer Blocks
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    # Global Average Pooling (取代 Flatten，保留更多特徵)
    x = layers.GlobalAveragePooling1D(data_format="channels_first")(x)

    # MLP Head
    for dim in mlp_units:
        x = layers.Dense(dim, activation="relu")(x)
        x = layers.Dropout(mlp_dropout)(x)

    outputs = layers.Dense(1, activation="linear")(x)

    model = models.Model(inputs, outputs, name="Transformer_Stock_Predictor")

    model.compile(
        loss="mse",
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        metrics=["mae"]
    )
    return model

# 建立模型
input_shape = (X_train.shape[1], X_train.shape[2])
model = build_transformer_model(
    input_shape,
    head_size=256,
    num_heads=4,
    ff_dim=4,
    num_transformer_blocks=4,
    mlp_units=[128],
    dropout=0.20,
    mlp_dropout=0.20
)

model.summary()

# [6] 模型訓練
callbacks = [tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)]

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,  # 依照助教參數
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# [7] 預測與評估
preds = model.predict(X_test)

# 反正規化
def inverse_transform(scaler, y_true, y_pred, n_features, target_idx):
    # 建立 dummy array 來做 inverse transform
    dummy_true = np.zeros((len(y_true), n_features))
    dummy_pred = np.zeros((len(y_pred), n_features))

    dummy_true[:, target_idx] = y_true.flatten()
    dummy_pred[:, target_idx] = y_pred.flatten()

    inv_true = scaler.inverse_transform(dummy_true)[:, target_idx]
    inv_pred = scaler.inverse_transform(dummy_pred)[:, target_idx]

    return inv_true, inv_pred

y_test_inv, preds_inv = inverse_transform(scaler, y_test, preds, len(feature_names), target_idx)

# 計算指標
mse = mean_squared_error(y_test_inv, preds_inv)
mae = mean_absolute_error(y_test_inv, preds_inv)
print(f"\nTransformer 測試集表現:")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")

# [8] 結果視覺化
plt.figure(figsize=(14, 7))
test_dates = stock_data.index[train_size + WINDOW_SIZE:]

# 確保長度一致
min_len = min(len(test_dates), len(y_test_inv))
plt.plot(test_dates[:min_len], y_test_inv[:min_len], label='Actual Price', color='black')
plt.plot(test_dates[:min_len], preds_inv[:min_len], label='Transformer Predicted', color='red', linestyle='--')

plt.title('聯發科 (2454) Transformer 股價預測結果')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 繪製 Loss 曲線
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training History')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
```
